# Data Loading and Description

```python
import pandas as pd
```

# Label Encode

```python
# encoding two "object labels" to "int"
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data['F23'])
data['F23'] = le.transform(data['F23'])

f23_class = le.classes_

print(f23_class)
```

```python
le.fit(data['F29'])
data['F29'] = le.transform(data['F29'])

f29_class = le.classes_

print(f29_class)
```

```python
# now all features are in int/float
data.info()
```

# Feature Selection

```python
# Create the correlation matrix
import matplotlib.pyplot as plt
import seaborn as sns

corr = data.corr()

plt.figure(figsize=(30, 20))

sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)

plt.title('Correlation Matrix')
plt.show()
```

```python
# selecting all features for further selection
features = list(data.columns[:34])
print("features:", features, sep="\n")
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

y = data["Target"]
X = data[features]

# selecting 22 best features using F statistic and p value
selector = SelectKBest(f_regression, k=22)
selector.fit(X, y)

# priniting 22 best selected features
X_new = selector.transform(X)
print(data.columns[selector.get_support(indices=True)].tolist())
final_features = data.columns[selector.get_support(indices=True)].tolist()
```

```python
# selecting 22 features and target variables
y = data["Target"]
X = data[final_features]
```

# Machine Learning Models

```python
# train test split in ratio 3:1
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# importing libraries for training models
import numpy as np
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import linear_model
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

## Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train)


pred= lr.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred))) 
print(r2_score(y_test, pred))
```

## Lasso Regression

```python
clf = linear_model.Lasso(alpha=10)
clf.fit(X_train, y_train)

```

```python
pred1 = clf.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred1))) 
print(r2_score(y_test, pred1))
```

## ElasticNet

```python
regr1 = ElasticNet(random_state=0, l1_ratio = 0.7)
regr1.fit(X_train, y_train)
```

```python
pred2 = regr1.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred2))) 
print(r2_score(y_test, pred2))
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regr2 = RandomForestRegressor(max_depth=200, random_state=0)
regr2.fit(X_train, y_train)

```

```python
pred3 = regr2.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred3))) 
print(r2_score(y_test, pred3))
```

```python
pipeline2 = Pipeline([('scaler', MinMaxScaler()),('regressor', RandomForestRegressor(max_depth=200, random_state=0))])

pipeline2.fit(X_train, y_train)

pred4 = pipeline2.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred4))) 
print(r2_score(y_test, pred4))
```

## Extra Trees

```python
reg3 = ExtraTreesRegressor(n_estimators=500, random_state=0).fit(X_train, y_train)
```

```python
pred5 = reg3.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred5))) 
print(r2_score(y_test, pred5))
```

```python
pipeline3 = Pipeline([('scaler', MinMaxScaler()), ('regressor', ExtraTreesRegressor(n_estimators=500, random_state=0))])

pipeline3.fit(X_train, y_train)

pred6 = pipeline3.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred6))) 
print(r2_score(y_test, pred6))
```

# Evaluating all the 7 models

```python
# making a df of results for easy interpretation and graphs
evaluation_dict = {'Linear Regression':[np.sqrt(mean_squared_error(y_test, pred)),r2_score(y_test, pred)], \
                   'Lasso Regression':[np.sqrt(mean_squared_error(y_test, pred1)),r2_score(y_test, pred1)], \
                   'Elastic Net':[np.sqrt(mean_squared_error(y_test, pred2)),r2_score(y_test, pred2)], \
                   'Random Forest':[np.sqrt(mean_squared_error(y_test, pred3)),r2_score(y_test, pred3)], \
                   'Random Forest with scaling':[np.sqrt(mean_squared_error(y_test, pred4)),r2_score(y_test, pred4)], \
                   'Extra Forest':[np.sqrt(mean_squared_error(y_test, pred5)),r2_score(y_test, pred5)], \
                   'Extra Forest with scaling':[np.sqrt(mean_squared_error(y_test, pred6)),r2_score(y_test, pred6)], \
                  }
```

```python
n = pd.DataFrame.from_dict(evaluation_dict, orient='index')
n = n.rename(columns={0: "RMSE", 1: "R2 Score"})
```

```python
n
```

```python
# plotting graphs

import matplotlib.pyplot as plt
import seaborn as sns

# Data for plotting
algorithms = list(n.index)
rmse = list(n["RMSE"])
r2_score = list(n["R2 Score"])

# Plot the bar chart for RMSE
plt.figure(figsize=(10, 6))
sns.barplot(x=algorithms, y=rmse, color='skyblue')
plt.xlabel('Algorithm')
plt.ylabel('RMSE')
plt.title('Comparison of RMSE for Different Algorithms')
plt.xticks(rotation=45)
plt.show()

# Plot the bar chart for R2 Score
plt.figure(figsize=(10, 6))
sns.barplot(x=algorithms, y=r2_score, color='lightgreen')
plt.xlabel('Algorithm')
plt.ylabel('R2 Score')
plt.title('Comparison of R2 Score for Different Algorithms')
plt.xticks(rotation=45)
plt.show()
```
